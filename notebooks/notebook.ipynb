```python
import pandas as pd
import numpy as np
```

```python

data = pd.read_csv('/content/drive/MyDrive/data/energy_dataset.csv')

cols = list(data.columns)
idx1 = cols.index('Energy_Production_MWh')
idx2 = cols.index('Jobs_Created')
cols[idx1], cols[idx2] = cols[idx2], cols[idx1]

data = data.sort_values(by='Jobs_Created')
data = data[cols]
pd.set_option('display.precision', 3)
print(data.columns)
print(f'')
```

```python
print(data['Type_of_Renewable_Energy'].unique())
print(data['Type_of_Renewable_Energy'].value_counts()[[1, 2]])
# print((data['Type_of_Renewable_Energy'].isin([1, 2]).sum()))
```

```python
items = [1, 2]
item_s = [1]
item_w = [2]

data_sw = data[data['Type_of_Renewable_Energy'].isin(items)]
data_s = data[data['Type_of_Renewable_Energy'].isin(item_s)]
data_w = data[data['Type_of_Renewable_Energy'].isin(item_w)]
print(data_sw.columns)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score, root_mean_squared_error, mean_squared_error, mean_absolute_error, accuracy_score, confusion_matrix
```

```python
x, y = data_sw.drop(columns=['Energy_Production_MWh', 'Type_of_Renewable_Energy', 'Jobs_Created', 'Funding_Sources', 'Financial_Incentives_USD']), data_sw.Energy_Production_MWh

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

#SCALING DOWN DATA
sc = StandardScaler()
x_train_scaled = sc.fit_transform(x_train)
x_test_scaled = sc.transform(x_test)

model = LinearRegression()
model.fit(x_train, y_train)

y_pred = model. predict(x_test)

rmse = root_mean_squared_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
r_score = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)

print(y_train.head()), print(x.columns)
print(f'rmse: {rmse}'), print(f'mse: {mse}'), print(f'r_score: {r_score}'), print(f'mae: {mae}'), print(f'coef: {model.coef_}')
```

```python
y_class = (y>250000).astype(int)

x_train, x_test, y_train, y_test = train_test_split(x, y_class, test_size=0.2, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)

model.fit(x_train, y_train)

y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
confusion = confusion_matrix(y_test, y_pred)
importances = model.feature_importances_

print(f'accuracy: {accuracy}'), print(f'confusion: {confusion}'), print(model.score(x_test, y_test)), print(f'importances: {importances}')
```
